# Histogram

```python
import cv2 as cv
import matplotlib.pyplot as plt

path = './lena.jpg'

lena = cv.imread(path, 0)
histogram = cv.equalizeHist(lena)

plt.figure(figsize = (12, 12))
plt.subplot(221)
plt.imshow(lena, cmap = 'gray'), plt.title('Original Gray')
plt.subplot(222)
plt.hist(lena), plt.title('Histogram Original Gray')
plt.subplot(223)
plt.imshow(histogram, cmap = 'gray'), plt.title('equalized')
plt.subplot(224)
plt.hist(histogram), plt.title('Histogram equalized')
plt.show()

```
